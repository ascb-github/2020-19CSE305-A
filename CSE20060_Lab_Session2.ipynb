```python
# A1) Tokenization: Load the documents for all the 37 students. Tokenize all the documents and store the tokens
df = pd.read_excel('Lab Session2 Data.xlsx')
```

```python
strings = []
df_array = np.array(df)
```

```python
for i in df_array[ : , -1]:
    strings.append(i)
```

```python
strings
```

```python
tokenized_words =[]
for i in strings:
    tokenized_words.append(word_tokenize(i))
```

```python
tokenized_words
```

```python
#A2. Token Population: Merging the tokens from all documents, create a master list of distinct tokens
#available across all documents. Let us call this as “token population”
entire_string = '\n'.join(strings)
print(entire_string)
```

```python
master_tokenized_words = word_tokenize(entire_string)
```

```python
master_tokenized_words
```

```python
# A3. Stop-words:
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
stop_words
# these words should be ignored while detecting plagiarism 
```

```python
# A4. Bag-of-Words:
bag_of_words_individual= []
  
for i in tokenized_words:
    list = []
    for w in i:
        if w not in stop_words:
            list.append(w)
    bag_of_words_individual.append(list)
```

```python
bag_of_words_individual
```

```python
bag_of_words_population = []
for i in tokenized_words:
    for w in i:
        if w not in stop_words:
            bag_of_words_population.append(w)
```

```python
bag_of_words_population
```

```python
# A5. Document Vectorization:
v1 = []
for i in range(len(tokenized_words)):
    temp = []
    for j in tokenized_words[i]:
        if j in bag_of_words_individual[i]:
            temp.append(1)
        else:
            temp.append(0)
    v1.append(temp)
```

```python
for i in v1:
    print(i)
```

```python
# creation of v2 - count of each word
v2 = []
for i in tokenized_words:
    temp = []
    for j in bag_of_words_population:
        temp.append(i.count(j))
    v2.append(temp)
```

```python
for i in v2:
    print(i)
```

```python
# A6. Similarity Measure: Calculate the Jaccard Coefficient between the document vectors. Use first vector for each document
# jaccard co-efficient"
def jaccard(a,b):
    f11 = 0
    f10 = 0
    f01 = 0
    small = len(a) if len(a) < len(b) else len(b)
    for i in range(small):
        if(a[i]==1 and b[i]==1):
            f11 = f11 + 1
        elif(a[i]==1 and b[i]==0):
            f10 = f10 + 1
        elif(a[i]==0 and b[i]==1):
            f01 = f01 + 1
    ans = f11/(f11+f01+f10) if f11+f01+f10 != 0 else 0
    return ( 0 if ans == 0 else ans)
jc_heatmap = []
for i in range(len(v1)):
    temp = []
    for j in range(len(v1)):
        print("The Jaccard Co efficient between {}th document and {}th document is {}".format(i,j,jaccard(v1[i],v1[j])))
        temp.append(jaccard(v1[i],v1[j]))
    jc_heatmap.append(temp)
    
```

```python
# A7. Calculate the Cosine similarity between the documents by using the second feature vector for each document.
def cosine_similarity(A, B):
    return np.dot(A,B)/(norm(A)*norm(B))
cs_heatmap =[]
for i in range(len(v2)):
    temp=[]
    for j in range(len(v2)):
        print("The Cosine Similarity between {}th document and {}th document is {}".format(i,j,cosine_similarity(v2[i],v2[j])))
        temp.append(cosine_similarity(v2[i],v2[j]))
    cs_heatmap.append(temp)
```

```python
# A8. Plot the Jaccard and cosine matrices and see the results.
# heatmap of Jaccard co efficient
map1 = sns.heatmap(data = jc_heatmap, cmap='Blues')
```

```python
# heatmap of Cosine Similarity
map1 = sns.heatmap(data = cs_heatmap, cmap='Reds')
```
